```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
summer = pd.read_csv('datasets\summer2016.csv')
```

<font size='5'> Bar Chart </font>

```python
sports = summer['Sport'].unique()
```

```python
fig, ax = plt.subplots()

for sport in sports[:11]:
    sport_df = summer[summer['Sport']==sport]
    ax.bar(sport, sport_df['Height'].mean(),
          yerr = sport_df['Height'].std())

ax.set_ylabel('Height (CM)')
ax.set_xticklabels(sports, rotation=90)
plt.show()
```
