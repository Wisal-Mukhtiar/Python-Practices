```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
medals = pd.read_csv('datasets\medals_by_country_2016.csv', index_col=0)
```

<font size='4'> Bar charts <font>

```python
fig, ax = plt.subplots()
ax.bar(medals.index, medals['Gold'])
```

names in above chart are really conjusted, let's rotate it by 90 degree

```python
# add ax.set_xticklabels
fig, ax = plt.subplots()
ax.bar(medals.index, medals['Gold'])
ax.set_xticklabels(medals.index, rotation=90)
```
