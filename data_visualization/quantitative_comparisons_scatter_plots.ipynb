```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
climate = pd.read_csv('datasets\climate_change.csv', index_col = 0)
```

<font size='5' > Scatter Plots </font>

```python
# scatter plot is called using scatter call method with axis 
fig, ax = plt.subplots()
ax.scatter(climate['co2'], climate['relative_temp'])
```

for selecting some specific time series, we can do it as 

```python
ninetees = climate['1980-01-01':'1989-12-31']
eightess = climate['1970-01-01':'1979-12-31']
```

```python
# we can add color argument as well as legend call 
fig, ax = plt.subplots()
ax.scatter(ninetees['co2'], ninetees['relative_temp'], label='90s')
ax.scatter(eightees['co2'], eightees['relative_temp'], label='80s')
```
