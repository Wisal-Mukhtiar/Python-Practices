```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
climate = pd.read_csv('datasets\climate_change.csv', parse_dates=['date'], index_col = 0)
```

<font size='5' > Scatter Plots </font>

```python
# scatter plot is called using scatter call method with axis 
fig, ax = plt.subplots()
ax.scatter(climate['co2'], climate['relative_temp'])
ax.set_xlabel('CO2 PPM')
ax.set_ylabel('relative temp')
```

for selecting some specific time series, we can do it as 

```python
ninetees = climate['1980-01-01':'1989-12-31']
eightees = climate['1970-01-01':'1979-12-31']
```

```python
# we can add color argument as well as legend call 
fig, ax = plt.subplots()
ax.scatter(ninetees['co2'], ninetees['relative_temp'], label='90s')
ax.scatter(eightees['co2'], eightees['relative_temp'], label='80s')
ax.set_xlabel('CO2 PPM')
ax.set_ylabel('relative temp')
ax.legend()
```

<font size='4'> Encoding third variable as color </font><br>
we will use the index as a color argument in the scatter plot

```python
# adding c method
fig, ax = plt.subplots()

ax.scatter(climate['co2'], climate['relative_temp'], c=climate.index)
ax.set_xlabel('CO2 PPM')
ax.set_ylabel('relative temp')
```

we see in the above chart that the coloring has been done accordingly with the dates
