```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
seattle_weather = pd.read_csv('datasets\seattle_weather.csv')
austin_weather = pd.read_csv('datasets\Austin_weather.csv')
```

<font size='5' > Simple line plots </font>

```python
fig, ax = plt.subplots()
ax.plot(seattle_weather['DATE'].head(10), seattle_weather['MLY-TAVG-NORMAL'].head(10))
ax.plot(austin_weather['DATE'].head(10), austin_weather['MLY-TAVG-NORMAL'].head(10))
plt.show()
```

<font size='4' > Customizing plot </font>


<font size='3' >adding marker, linestyle and color to the lines </font>

```python
fig, ax = plt.subplots()
ax.plot (seattle_weather['DATE'].head(10), seattle_weather['MLY-TAVG-NORMAL'].head(10),
        marker='o', linestyle='--', color='r')
plt.show()
```

<font size='3' >setting labels and titles </font>

```python
fig, ax = plt.subplots()
ax.plot (seattle_weather['DATE'].head(10), seattle_weather['MLY-TAVG-NORMAL'].head(10),
        marker='o', linestyle='--', color='r')
ax.set_xlabel('Date')
ax.set_ylabel('Temperature')
ax.set_title('Temperature Vs Date')
plt.show()
```

<font size='5'> Multiple Subplots </font>

```python
fig, ax = plt.subplots(2,1) # creating two rows, with one column (2 subplots in one column)
ax[0].plot(seattle_weather['DATE'].head(10), seattle_weather['MLY-PRCP-NORMAL'].head(10), c='b', marker='o')
ax[1].plot(austin_weather['DATE'].head(10), austin_weather['MLY-PRCP-NORMAL'].head(10), c='r', marker='x')
plt.show()
```
