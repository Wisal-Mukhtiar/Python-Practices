```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
summers = pd.read_csv('datasets\summer2016.csv')
```

```python
rowing_male = summers[(summers['Sport']=='Rowing') & (summers['Sex']=='M')].head(30)
gymnast_male = summers[(summers['Sport']=='Gymnastics') & (summers['Sex']=='M')].head(30)
```

<font size='4'> Histograms </font>

```python
fig, ax = plt.subplots()

ax.hist(rowing_male['Weight'])

ax.hist(gymnast_male['Weight'])

ax.set_xlabel('Weight')
ax.set_ylabel('# of observations')

plt.show()
```

adding number of bins and also the histogram type

```python
fig, ax = plt.subplots()

#histtype and bins arg added
ax.hist(rowing_male['Weight'], histtype='step', bins=4)

ax.hist(gymnast_male['Weight'], histtype='step', bins=4)

ax.set_xlabel('Weight')
ax.set_ylabel('# of observations')

plt.show()
```

setting labels and calling legend

```python
fig, ax = plt.subplots()

#label arg and legend call added
ax.hist(rowing_male['Weight'], histtype='step', bins=4, label='Rowing')

ax.hist(gymnast_male['Weight'], histtype='step', bins=4, label='Gymnast')

ax.set_xlabel('Weight')
ax.set_ylabel('# of observations')

ax.legend()

plt.show()
```
