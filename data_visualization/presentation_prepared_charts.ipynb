```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
seattle_weather = pd.read_csv('datasets\seattle_weather.csv')
austin_weather = pd.read_csv('datasets\Austin_weather.csv')
```

<font size='5'> Choosing style of your figure </font>

```python
# we use plt.style.use('default') option, by default, we can change it to many other styles e.g
plt.style.use('ggplot')
fig, ax = plt.subplots()
ax.plot(seattle_weather['DATE'].head(10), seattle_weather['MLY-TAVG-NORMAL'].head(10))
ax.plot(austin_weather['DATE'].head(10), austin_weather['MLY-TAVG-NORMAL'].head(10))
ax.set_xlabel('Date')
ax.set_ylabel('Temperature')
ax.set_title('Temperature Vs Date')
plt.show()
```
